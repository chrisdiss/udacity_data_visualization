There is a huge difference in the amount of started and ended trips per station, which indicated that some stations are heavily used whereas others not. To investigate the stations closer, we take a detailed look on stations with counts > 1000.

```python
plt.figure(figsize=(20,10))

freq = df['start_station_id'].value_counts().loc[lambda x: x>1000]
gen_order = freq.index
sb.countplot( data=df, x='start_station_id', color=base_color, order=gen_order);
```

```python
plt.figure(figsize=(20,10))

freq = df['end_station_id'].value_counts().loc[lambda x: x>1000]
gen_order = freq.index
sb.countplot( data=df, x='end_station_id', color=base_color, order=gen_order);
```

For better analyses I create a two new dataframes with all the relevant information that can be linked to a special station. First there will be a dataframe that contains all the information about every trip related to a station. The second dataframe will contain all the information about a single station.

```python
# create a dataframe that holds the information about how many trips ended at specific station
df_stations = df['end_station_id'].value_counts()
df_stations = df_stations.to_frame().reset_index()
df_stations = df_stations.rename(columns= {'index': 'station_id', 'end_station_id': 'end_counts'})
df_stations.index.station_id = 'index'
```

```python
# create a dataframe that holds information about how many trips started at a specific station and merge it with the information about the ending points
df_stations_1 = df['start_station_id'].value_counts()
df_stations_1 = df_stations_1.to_frame().reset_index()
df_stations_1 = df_stations_1.rename(columns= {'index': 'station_id', 'start_station_id': 'start_counts'})
df_stations = pd.merge(df_stations_1, df_stations, on=['station_id'])
df_stations.head()
```

```python
# create a dataframe with the coordinates and at it to an all information dataframe which makes it possible to see from which station the trips started and where they ended
df_coordinates = df.copy()
df_coordinates.drop(['duration_sec', 'start_time', 'end_time', 'bike_id', 'user_type', 'member_birth_year', 'member_gender', 'bike_share_for_all_trip', 'duration_min'], axis=1, inplace=True)
df_coordinates = df_coordinates.rename(columns= {'start_station_id': 'station_id'})
df_stations_all = pd.merge(df_stations, df_coordinates, on=['station_id'])
df_stations_all.head()
```

```python
# create a dataframe with the overview information about one single station
df_coordinates.drop(['end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude'], axis=1, inplace=True)
df_stations_overview = pd.merge(df_stations, df_coordinates, on=['station_id'])
df_stations_overview.drop_duplicates(inplace=True)
df_stations_overview.head()
```

## Bivariate Analysis

##### Stations information

```python
# How are starting and endpoint stations connected?
ticks = [10,50,100,500,1000,5000]
labels = ['{}'.format(v) for v in ticks]

plt.scatter(data = df_stations_overview, x = 'start_counts', y = 'end_counts', alpha= 1/5);
plt.xlabel('Count of starting points');
plt.ylabel('Count of ending points');

plt.xscale('log');
plt.xticks(ticks, labels);
plt.yscale('log');
plt.yticks(ticks, labels);
```

There is a clear correlation between number of starting and ending trips per station. This indicates that there are stations that generally more frequented than others.

```python
# Are there some bikes connected to some stations?
plt.scatter(data = df, x = 'start_station_id', y = 'bike_id', alpha = 1/100)
plt.xlabel('Start Station ID')
plt.ylabel('Bike ID');
```

The data shows that a special bike ist likely to not being used more frequently at a certain station. 

```python
# investigate the connections between stations
plt.figure(figsize=(15,15))
plt.scatter(data = df, x = 'start_station_id', y = 'end_station_id', alpha= 1/20);
plt.xlabel('Starting station ID');
plt.ylabel('Ending station ID');
```

The data shows that there are some clusters of stations that are highly connected. E.g. stations with the ID between 0 and 150 and around 350 seem to be connected within each other. This means that a trip starting at one of these stations will probably end at a station within the range of stations. Probably this is due  the stations are located in a similar area. Especially stations around ID 250 seem to be highly connected with each other. Also interestingly some stations (e.g. around ID 325) have a high frequency and interconnection to other clusters of stations which makes them  potentialy interesting stations.

##### User behavior

```python
# Does the duration of single trips (user behavior) vary depending on the day time?
plt.figure(figsize=(15,5))
sb.violinplot(data=df, x=df.start_time.dt.hour, y='duration_min');
plt.xlim(4,22);
plt.ylim(0,40);
```

The data shows that generally spoken there is probably no difference in the duration of trips depending on the start time.

##### User characteristica

```python
# Can we spot a difference in some user characteristics?
sb.countplot(data = df, x = 'user_type', hue = 'member_gender');
```

Even the ratio from male / female in the subscriber count seems a bit higher (and it is), this difference seems to be not crucial as the data shows that in both cases customers and subscribers the male gender is more spread.

```python
# is there some difference in behavior based on user type and gender?
df.groupby(['user_type','member_gender']).mean()
```

```python
counts = df.groupby(['user_type','member_gender']).duration_min.mean()
counts = counts.reset_index(name='duration_min')
counts
```

```python
counts = counts.pivot(index = 'user_type', columns = 'member_gender', values = 'duration_min')
```

```python
sb.heatmap(counts)
```

It can be clearly seen that users from type customers do have a higher mean durations of their trip in all three categories (female, male, other).

```python
# investigate the different trip durations per user type in a violinplot
plt.figure(figsize=(10,5))
sb.violinplot(data=df, x='user_type', y='duration_min');
plt.ylim(0,50);
```

The plots show that also the median of trip duration lies higher in customers than in subscribers. Generally spoken, subscribers tend to have shorter trips whereas customers have longer trips.

```python
# investigate if the user behavior (duration) depends on the gender variable
plt.figure(figsize=(10,5))
sb.violinplot(data=df, x='member_gender', y='duration_min');
plt.ylim(0,50);
```

The data shows that gender does not really inlfuence the duration of the trips.

```python
# investigate if the age influences the behavior
plt.figure(figsize=(10,5))
plt.scatter(data = df, x = 'member_birth_year', y = 'duration_min', alpha= 1/20);
plt.xlabel('Members birth year');
plt.ylabel('Duration in minutes');
plt.xlim(1930,2005);
plt.ylim(0,50)
```

Again, it seems that age does not really influence the duration of trips also by zooming into the data it can be seen that there are some slightly higher duration values for the more younger people.

```python
# investigate if there are some stations that are used by more younger or older people
plt.figure(figsize=(10,5))
plt.scatter(data = df, x = 'member_birth_year', y = 'start_station_id', alpha= 1/20);
plt.xlabel('Members birth year');
plt.ylabel('Start station ID');
plt.xlim(1930,2005);
```

Data shows that there are no stations that can be systematically connected with different age populations. Only stations with the ID around 300 could have a younger audience.

```python
# closer look at stations with ID around 300
plt.figure(figsize=(10,5))
plt.scatter(data = df, x = 'member_birth_year', y = 'start_station_id', alpha= 1/5);
plt.xlabel('Members birth year');
plt.ylabel('Start station ID');
plt.xlim(1930,2005);
plt.ylim(290,310);
```

Also the data shows higher usage by younger people, this seems to be unsystematic and also connected to the smaller usage frequency in average.

```python
# investigate if the user type is different between the stations
plt.figure(figsize=(10,5))
sb.violinplot(data=df, x='user_type', y='start_station_id');
plt.ylim(0,500);
```

Although different in absolut values, the shape of the plot indicates that for the user types there is no difference in start stations.

```python
# investigate if the user types varies on the age of the members
plt.figure(figsize=(10,5))
sb.violinplot(data=df, x='user_type', y='member_birth_year');
plt.ylim(1930,2010);
```

The distribution of user types depending on the age is not different

```python
# investiagte if gender and age is connected
plt.figure(figsize=(10,5))
sb.violinplot(data=df, x='member_gender', y='member_birth_year');
plt.ylim(1930,2010);
```

The shape of, especially female and male is quite similar which shows that the distribution of members based on gender and age is similar.

##### Conclusions
After inspecting the bivariate analysis there are some insights we can use to drive multivariate analysis. From the perspective of the bike sharing service, there are some stations that are highly frequented and others not. Even on a daily and hourly bases the usage pattern differs. This information can be used to improve the bike sharing service. Also, it seems that there is a difference in usage behavior (trip duration) between the the types customer and subscriber. Other user variables do not seem to influence the behavior, despite the fact that there are much more male users than female.
